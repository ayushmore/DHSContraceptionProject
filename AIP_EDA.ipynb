```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AIP_DATA/india_couples_dataset.csv")
df.head()
```

```python
data_vars = pd.read_csv("AIP_DATA/AIP_VARIABLES.csv")
data_vars.columns = ["Variables", "Definition", "Include"]
key_vals = data_vars[data_vars["Include"] == 1]
key_vals[key_vals["Variables"] == "v312"]
```

```python
variable_col = key_vals["Variables"].tolist()
description_col = key_vals["Definition"].tolist()

x = key_vals[key_vals["Variables"].str.contains("v307")]
v307_names = ["Pill", "IUD", "Inj.", "Dia.", "M.C.", "F.S.", "M.S.", "P.A.", "With.", "O.T.", "Imp.", "Abs.", "Lac.", "F.C", "F&J.", "E.C.", "O.M.", "STM", "CS1", "CS2"]
v307_names = ["Method Currently Used: " + i for i in v307_names]

v307_idx = 0
for i in range(len(description_col)):
    if description_col[i] == "Method currently used":
        description_col[i] = v307_names[v307_idx]
        v307_idx += 1

clean_data_desc = df.filter(items = variable_col)
clean_data_desc.columns = description_col
clean_data_desc

clean_data_labels = df.filter(items = variable_col)
#clean_data_labels
clean_data_desc
```

```python
nulls = clean_data_desc.isnull().mean().sort_values(ascending = False)
new_nulls = clean_data_labels.isnull().mean().sort_values(ascending = False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(nulls)

# (nulls < 0.75).value_counts()
# nulls
```
