```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AIP_DATA/india_couples_dataset.csv")
df.head()
```

```python
data_vars = pd.read_csv("AIP_DATA/AIP_VARIABLES.csv")
data_vars.columns = ["Variables", "Definition", "Include"]
key_vals = data_vars[data_vars["Include"] == 1]
key_vals[key_vals["Variables"] == "v312"]

outcome_categories = pd.read_csv("AIP_DATA/Outcome_categories.csv")
outcome_categories["Label"] = "Method Currently Used: " + outcome_categories["Label"]
outcome_categories

adjust_outcome = pd.read_csv("AIP_DATA/Outcome_categories.csv")
rename_labels = {"Not Using Any Method": "not using", "Male Condom": "condom", 
        "Periodic Abstinense (Rhythm)": "rhythm/periodic abstinence", "Male Strilization": "male sterilization",
        "Lactational ammenorhea": "lactational amenorrhea (lam)", "Foam and Jelly": "Foam or Jelly",
        "Standard Days method": "standard days", "Withdrawl": "withdrawal"}
adjust_outcome["Label"] = adjust_outcome["Label"].replace(rename_labels).str.lower()
adjust_outcome = adjust_outcome.drop(columns = ["Code"])
adjust_outcome = adjust_outcome.rename(columns = {"Label": "Current contraceptive method"})
adjust_outcome
```

```python
variable_col = key_vals["Variables"].tolist()
description_col = key_vals["Definition"].tolist()

x = key_vals[key_vals["Variables"].str.contains("v307")]
v307_names = ["Pill", "IUD", "Inj.", "Dia.", "M.C.", "F.S.", "M.S.", "P.A.", "With.", "O.T.", "Imp.", "Abs.", "Lac.", "F.C", "F&J.", "E.C.", "O.M.", "STM", "CS1", "CS2"]
v307_names = ["Method Currently Used: " + i for i in v307_names]

labels = outcome_categories["Label"].tolist()

l_idx = 0
for i in range(len(description_col)):
    if description_col[i] == "Method currently used":
        description_col[i] = labels[l_idx]
        l_idx += 1

clean_data_desc = df.filter(items = variable_col)
clean_data_desc.columns = description_col
clean_data_desc

clean_data_labels = df.filter(items = variable_col)
#clean_data_labels
clean_data_desc.head()
```

```python
comb = pd.merge(clean_data_desc, adjust_outcome, on = 'Current contraceptive method', how="left")
n = comb.isnull().mean().sort_values(ascending = False)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(n)
```

```python
clean_comb = comb[comb.columns[comb.isnull().mean() < 0.75]]
compression_opts = dict(method='zip',
                        archive_name='clean_data.csv')  
clean_comb.to_csv('clean_data.zip', index=False,
          compression=compression_opts)
```
